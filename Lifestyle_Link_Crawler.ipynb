```
# Create a python file inside the /content/drive/MyDrive/Colab Notebooks/crawler_tutorial/crawler_tutorial/spiders folder and put this code in there

import scrapy

class WebSpider(scrapy.Spider):
    name = "lifestyle"

    start_urls = [ 
                   "https://www.primermagazine.com/"
                   ]
    def parse(self, response):
        yield {
          "link":response.css("div.category-table a::attr(href)").extract()
          }
        # follow pagination
        next_page_url = response.xpath("//a[@class='next-page']/@href").extract_first()
        if next_page_url:
            yield scrapy.Request(url=next_page_url, callback=self.parse)
```

```
import pandas as pd
```

```
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Link_Crawler/links.csv") 
```

```
df.head()
```

```
list_1 = []
list_of_urls = []

for ind in df.index: 
     list_1.append(df['link'][ind]) 
     for links in list_1:
        links_split = links.split(',')
        for link in links_split:
            list_of_urls.append(link)

list_of_urls = list(set(list_of_urls))
```

```
print(f"The number of links is {len(list_of_urls)}")
```

```
!pip install newspaper3k

from newspaper import Article
import nltk
nltk.download('punkt')
```

```
seperator = ', '

def converttostr(input_seq, seperator):
    final_str = seperator.join(input_seq)
    return final_str

for ind, url in enumerate(list_of_urls, start=1):
    article = Article(url, language="en")

    article.download() 
    article.parse() 
    article.nlp() 

    
    with open("Guardian Technology File.txt", "a") as f:
        f.write(article.title + "\n" + "\n")
        f.write(article.text + "\n" + "\n" + "\n")
        f.write(f"This is the article summary \n {article.summary}" + "\n" + "\n")
        keywords = converttostr(article.keywords, seperator)
        f.write(f"These are the keywords of this article [{keywords}]" + "\n" + "\n" + "\n" + "\n" + "\n")

    print(f"Finish with article {ind}")
```
